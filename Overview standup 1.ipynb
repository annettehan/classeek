### 4. Merge Course Schedule and Descriptions

```python
def merge_df(open_df, des_df):
    
    # Create new column of course subject i.e. AAS 103 in open_df
    subject = open_df['Subject']
    open_df['subject'] = [s[s.find("(")+1:s.find(")")] for s in subject]
    open_df['course'] = open_df['subject'] + open_df['Catalog Nbr']
    open_df.sort_values(by=['course'], inplace=True)
    
    # Create new column of course subject i.e. AAS 103 in des_df
    des_df['Catalog Nbr'] = des_df['catalog_number'].dropna().astype(int).astype(str)
    des_df['course'] = des_df['subject'] + ' ' + des_df['Catalog Nbr']
    
    # Merge these dfs
    df = open_df.merge(des_df, on=['course'], how='left', copy=False)
    
    cols = ['Class Nbr', 'course', 'Term', 'Session', 'Acad Group', 'Subject', 'Course Title', 'description', 'Component', 'Time', 'Location', 
        'Instructor', 'Total', 'Seats Remaining', 'Has WL', 'Units', 'sub_title', 'credits', 'requirements_distribution', 'consent', 
        'advisory_prerequisites', 'other_course_info', 'repeatability']
    
    df = df[cols]
    
    return df
```

```python
w_22_merge = merge_df(w_22_open, w_22_des)
```

```python
f_21_merge = merge_df(f_21_open, f_21_des)
```

```python
f_21_merge.to_csv('f_21_merge.csv', index=False)
```

```python
w_22_merge.to_csv('w_22_merge.csv', index=False)
```
