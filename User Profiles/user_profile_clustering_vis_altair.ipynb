```python
tsne_df = pd.read_csv('student_profiles_chart_df.csv')
```

```python
selection = alt.selection_single(fields=['clusters'], on='click', bind='legend')
```

```python
color_range = ["#f3ad00",
"#ff6706",
"#0cffe9",
"#ef6bf6",
"#c80707",
"#2c2ca9",
"#f1e885",
"#3d8516",
"#533474",
"#f4677b",
"#8c377f",
"#3c7974",
"#ff9200",
"#d32aac",
"#96d66d",
"#ff5f16",
"#e8d407",
"#3fbe70",
"#c269fe",
"#1fa1d2",
"#003c71",
"#f84d9b"]
```

```python
color = alt.condition(selection,
                      alt.Color('clusters:N', title='Cluster', scale=alt.Scale(range=color_range)),
                      alt.value('lightgray'))
```

```python
chart = alt.Chart(tsne_df).mark_circle(opacity=0.5, size=65).encode(
            x=alt.X('Dim #1'),
            y=alt.Y('Dim #2'),
            color=color

).properties(width=900, height=600, title=
             {'text': ["Student User Profile Clusters"],
           'subtitle': ["Based on Students' Prior Coursework"]}

             
).configure_axis(domain=False, labelPadding=5, title=None, ticks=False, labels=False

).add_selection(
        selection
    
).configure_title(
    fontSize=20,
    subtitleFontSize=15,
    fontWeight='bold',
    subtitleFontWeight='bolder',
    subtitlePadding=7,
    offset=16
    
).configure_legend(
    strokeColor='gray',
    fillColor='#FCFCFC',
    padding=20,
    cornerRadius=15,
    orient='bottom-left',
    columns=2,
    titleFontSize=12,
    titleAnchor='middle'
)
```
